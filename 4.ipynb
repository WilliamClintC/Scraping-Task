```python
import os
import random
import fitz  # PyMuPDF

# Step 1: Define the folder path and pick 5 random PDF files
pdf_folder = r"C:\Users\clint\Desktop\Scraping Task\pdfs"
image_output_folder = os.path.join(pdf_folder, "Images")
os.makedirs(image_output_folder, exist_ok=True)

pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

if len(pdf_files) < 5:
    print("Not enough PDF files in the folder.")
    exit()

random_pdfs = random.sample(pdf_files, 5)

# Step 2: Extract the chart (render pages and save as images)
for pdf_file in random_pdfs:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    doc = fitz.open(pdf_path)
    print(f"Processing: {pdf_file}")

    for page_num in range(len(doc)):
        page = doc[page_num]
        pix = page.get_pixmap()  # Render the page as an image
        output_image_path = os.path.join(image_output_folder, f"{os.path.splitext(pdf_file)[0]}_page{page_num + 1}.png")
        pix.save(output_image_path)
        print(f"Saved page {page_num + 1} as image: {output_image_path}")

    # Here, you would add logic to identify and extract the specific chart
    # This could involve using OpenCV to detect the chart based on its visual properties

    doc.close()

print("Extraction complete. Check the 'Images' folder.")
```
