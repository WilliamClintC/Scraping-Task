```python
import pandas as pd

# Load the CSV file
file_path = r"C:\Users\clint\Desktop\Scraping Task\pdf_image_data.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data
```

```python
import numpy as np

# Create a function to extract month and year from different filename formats
def extract_date_components(filename):
    # Remove .pdf extension if present
    if '.pdf' in filename:
        filename = filename.replace('.pdf', '')
    
    # Check if the filename contains an underscore
    if '_' in filename:
        parts = filename.split('_')
        if len(parts) >= 2:
            month, year = parts[0], parts[1]
        else:
            month, year = np.nan, parts[0]
    else:
        # If no underscore, assume the entire string is the year
        month, year = 1, filename  # Default to January if no month
    
    # Return as numeric values, handling conversion errors
    try:
        month = pd.to_numeric(month)
    except:
        month = np.nan
        
    try:
        year = pd.to_numeric(year)
    except:
        year = np.nan
        
    return month, year

# Apply the function to extract date components
data['month'] = data['pdf_filename'].apply(lambda x: extract_date_components(x)[0])
data['year'] = data['pdf_filename'].apply(lambda x: extract_date_components(x)[1])

# Create datetime column with error handling
data['date'] = pd.to_datetime({'year': data['year'], 'month': data['month'], 'day': 1}, errors='coerce')

data
```

```python
data.loc[data['accepted_images_count'] == 0, 'note'] = 'Manual correction implemented'
data
```

```python
# Sort the data by the 'date_extracted' column in ascending order
data['date_extracted'] = pd.to_datetime(data['date'])  # Ensure the column is in datetime format
data = data.sort_values(by='date_extracted', ascending=True)

data
```
